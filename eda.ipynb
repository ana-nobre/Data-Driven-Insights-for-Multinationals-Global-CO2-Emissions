## 📊 Understanding the Descriptive Statistics Output
- **count**  
  Number of non-missing (non-null) values in that column.  
- **mean**  
  The arithmetic average of all values in the column.  
- **std (standard deviation)**  
  A measure of how much values vary around the mean  → High `std` means the values are very spread out, low `std` means they are clustered close to the mean.
- **min**  
  The smallest value observed in the column → Useful for spotting the lower boundary (e.g., countries with very small populations or GDPs).
- **25% (first quartile / Q1)**  
  The value below which 25% of the data falls → Helps understand the lower part of the distribution.
- **50% (median / Q2)**  
  The middle value when the data is sorted → Less sensitive to outliers than the mean.
- **75% (third quartile / Q3)**  
  The value below which 75% of the data falls → Helps understand the upper part of the distribution.
- **max**  
  The largest value observed in the column → Useful for identifying outliers or extreme cases (e.g., China or USA in emissions or GDP).

## 📑 Explanation of Key Variables in the Dataset

- **gdp**  
  Gross Domestic Product (GDP), usually measured in constant international dollars.

- **cement_co2**  
  CO₂ emissions from cement production (absolute values).

- **cement_co2_per_capita**  
  CO₂ emissions from cement production per person (tons per capita).

- **temperature_change_from_n2o**  
  Estimated temperature change contribution from nitrous oxide (N₂O), measured in °C.

- **total_ghg**  
  Total greenhouse gas emissions, measured in CO₂-equivalent (includes CO₂, CH₄, N₂O, etc.).

- **total_ghg_excluding_lucf**  
  Total greenhouse gas emissions excluding land use change and forestry (LUCF).

- **trade_co2**  
  CO₂ emissions embedded in traded goods and services (can be negative if a country is a net importer of embodied emissions).

- **trade_co2_share**  
  Share (%) of trade-related CO₂ emissions relative to total CO₂ emissions.

## 📌 Key Patterns Identified

- **Strong global inequality in emissions**  
  → From `total_ghg`: the **mean = 249.37** is far above the **median = 47.07**, and the **max = 13,968.92**, indicating a few extreme emitters (e.g., China, US, India) dominate totals while most countries are much smaller.

- **Economic size and population as drivers of emissions**  
  → `population`: **mean = 32,928,978.93** vs **median = 5,895,869.50**.  
  → `gdp`: **mean = 571,703,496,695.95** vs **median = 75,607,248,896.00**.  
  These gaps show that a small number of very large economies/populations skew global averages, linking economic and demographic size to emission levels.

- **Median GHG emissions much lower than the mean**  
  → Again in `total_ghg`: **median = 47.07** vs **mean = 249.37**, confirming a skewed distribution where most countries emit relatively little while a few contribute disproportionately.

- **Fossil fuels dominate, cement is secondary**  
  → `cement_co2` has **max = 849.98** (and `cement_co2_per_capita` **max = 1.24**), which is modest compared to fossil-fuel emissions columns, underscoring the dominance of coal, oil, and gas while cement remains smaller but relevant.

- **Trade shifts responsibility for emissions**  
  → `trade_co2`: **min = −1,532.08**, **max = 654.14**; `trade_co2_share`: **min = −98.85**, **max = 568.63**.  
  Negative values indicate **net importers** of embodied CO₂, positive values indicate **net exporters**, revealing asymmetry between production- and consumption-based emissions.
