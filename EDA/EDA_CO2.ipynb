## Outliers

```python
import seaborn as sns
```

```python
# last available year per country

df_latest = (df_country.sort_values(["country", "year"])     # sort to ensure "last row" is the latest year
              .groupby("country", as_index=False).tail(1))                              # keep only the latest year per country


topN = 50  #cohort size (50 biggest emitters)
```

```python
# GLOBAL — TOTAL CO2 (Top-N + IQR)
df_top_total = df_latest.sort_values("co2", ascending=False).head(topN).copy()

# IQR on this focused cohort
q1 = df_top_total["co2"].quantile(0.25)            # 25th percentile
q3 = df_top_total["co2"].quantile(0.75)            # 75th percentile
iqr = q3 - q1                                      # interquartile range
lo, hi = q1 - 1.5*iqr, q3 + 1.5*iqr                

df_top_total["out_total_top_iqr"] = (df_top_total["co2"] < lo) | (df_top_total["co2"] > hi)

print("=== GLOBAL — TOTAL CO2 — TopN + IQR thresholds ===")
print({"Q1": q1, "Q3": q3, "IQR": iqr, "LOWER": lo, "UPPER": hi})

outliers_total_top_iqr = (df_top_total.loc[df_top_total["out_total_top_iqr"], ["country", "Continent", "year", "co2"]]
                 .sort_values("co2", ascending=False))
outliers_total_top_iqr.head(20)   # remove .head(20) to see all
```

```python
# GLOBAL — CO2 PER CAPITA (Top-N + IQR)

df_top_pc = df_latest.sort_values("co2 per capita", ascending=False).head(topN).copy()

q1_pc = df_top_pc["co2 per capita"].quantile(0.25)
q3_pc = df_top_pc["co2 per capita"].quantile(0.75)
iqr_pc = q3_pc - q1_pc
lo_pc, hi_pc = q1_pc - 1.5*iqr_pc, q3_pc + 1.5*iqr_pc

df_top_pc["out_pc_top_iqr"] = (df_top_pc["co2 per capita"] < lo_pc) | (df_top_pc["co2 per capita"] > hi_pc)

print("=== GLOBAL — CO2 PER CAPITA — TopN + IQR thresholds ===")
print({"Q1": q1_pc, "Q3": q3_pc, "IQR": iqr_pc, "LOWER": lo_pc, "UPPER": hi_pc})

outliers_pc_top_iqr = (
    df_top_pc.loc[df_top_pc["out_pc_top_iqr"], ["country", "Continent", "year", "co2 per capita"]]
             .sort_values("co2 per capita", ascending=False)
)
outliers_pc_top_iqr.head(20)
```

```python
# For each continent, Top-N by total CO2, IQR per continent.
def iqr_flags(s):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    lo, hi = q1 - 1.5*iqr, q3 + 1.5*iqr
    return (s < lo) | (s > hi)

# Top-N per continent
df_cont_top_total = (
    df_latest.sort_values(["Continent", "co2"], ascending=[True, False]).groupby("Continent", as_index=False).head(topN).copy())

df_cont_top_total["out_total_top_iqr_cont"] = (df_cont_top_total.groupby("Continent")["co2"].transform(iqr_flags))

outliers_total_top_iqr_cont = (df_cont_top_total.loc[df_cont_top_total["out_total_top_iqr_cont"],
                          ["country", "Continent", "year", "co2"]].sort_values(["Continent", "co2"], ascending=[True, False]))

outliers_total_top_iqr_cont.head(30)
```

```python
# BY-CONTINENT — CO2 PER CAPITA (Top-N within each continent)

df_cont_top_pc = (df_latest.sort_values(["Continent", "co2 per capita"], ascending=[True, False]).groupby("Continent", as_index=False).head(topN).copy())

df_cont_top_pc["out_pc_top_iqr_cont"] = (df_cont_top_pc.groupby("Continent")["co2 per capita"].transform(iqr_flags))

outliers_pc_top_iqr_cont = (df_cont_top_pc.loc[df_cont_top_pc["out_pc_top_iqr_cont"],["country", "Continent", "year", "co2 per capita"]].sort_values(["Continent", "co2 per capita"], ascending=[True, False]))

outliers_pc_top_iqr_cont.head(30)
```
