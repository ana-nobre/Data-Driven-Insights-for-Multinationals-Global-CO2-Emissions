```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_country = pd.read_csv("data_clean_country.csv")
df_regions = pd.read_csv("data_clean_regions.csv")
```

```python
df_country.head()
```

```python
df_regions.head()
```

```python

# ==========================
# 1. Evaluation by country
# ==========================
top_countries = (df_country.groupby("Name")["co2"].sum().sort_values(ascending=False).head(10))
print("Top 10 COâ‚‚ emitting countries (historical):\n", top_countries)

top_countries.plot(kind="bar", figsize=(8,4), title="Top 10 COâ‚‚ emitting countries")
plt.ylabel("Total COâ‚‚")
plt.show()

```

```python

# ==========================
# 2. Evaluation by region (continent)
# ==========================
regions = (
    df_regions.groupby("Name")["co2"].sum().sort_values(ascending=False))
print("\nTotal COâ‚‚ emissions by continent/region:\n", regions)

regions.plot(kind="bar", figsize=(7,4), title="Total COâ‚‚ emissions by continent/region")
plt.ylabel("Total COâ‚‚")
plt.show()
```

```python

# ==========================
# 3. Evaluation by sector (country-level data)
# ==========================
sectors = ["coal_co2", "oil_co2", "gas_co2", "cement_co2", "land_use_change_co2"]
existing_sectors = [s for s in sectors if s in df_country.columns]

if existing_sectors:
    total_sectors = df_country[existing_sectors].sum().sort_values(ascending=False)
    print("\nTotal contribution by sector (global sum):\n", total_sectors)

    total_sectors.plot(kind="bar", figsize=(7,4), title="Global COâ‚‚ contribution by sector")
    plt.ylabel("Total COâ‚‚")
    plt.show()
```
